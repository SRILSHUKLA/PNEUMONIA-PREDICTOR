```python
import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split
```

```python


# Define paths
data_dir = "data/"
output_dir = "split_data/"
os.makedirs(output_dir, exist_ok=True)

# Class names (must match folder names in 'data/')
classes = ["covid", "normal", "pneumonia"]

# Create train/test directories
for split in ["train", "test"]:
    for class_name in classes:
        os.makedirs(os.path.join(output_dir, split, class_name), exist_ok=True)

# Split ratio (80% train, 20% test)
test_size = 0.2

# Split images into train/test for each class
for class_name in classes:
    # Get all image paths for the current class
    class_dir = os.path.join(data_dir, class_name)
    images = [f for f in os.listdir(class_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
    
    # Split into train/test
    train_imgs, test_imgs = train_test_split(images, test_size=test_size, random_state=42)
    
    # Copy images to train directory
    for img in train_imgs:
        src = os.path.join(class_dir, img)
        dst = os.path.join(output_dir, "train", class_name, img)
        shutil.copy(src, dst)
    
    # Copy images to test directory
    for img in test_imgs:
        src = os.path.join(class_dir, img)
        dst = os.path.join(output_dir, "test", class_name, img)
        shutil.copy(src, dst)

print("Dataset split completed!")
print(f"Train/test directories created at: {output_dir}")
```
